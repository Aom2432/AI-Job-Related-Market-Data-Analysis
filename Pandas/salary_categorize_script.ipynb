```python
import pandas as pd
import matplotlib as pyplot
```

```python
salary_skills = pd.read_csv("salary_by_skills.csv").drop(columns = ["ID_cleaned"])
salary_skills
```

```python
category_salary = pd.DataFrame(columns = ["skill_name", "salary_usd"])
category_salary
```

```python
is_AWS = salary_skills["AWS"].dropna()
category_salary["salary_usd"] = is_AWS
category_salary["skill_name"] = "AWS"
category_salary
```

```python
# Create new dataframe for each skills
azure = pd.DataFrame(columns = ["skill_name", "salary_usd"])
docker = pd.DataFrame(columns = ["skill_name", "salary_usd"])
GCP = pd.DataFrame(columns = ["skill_name", "salary_usd"])
Git = pd.DataFrame(columns = ["skill_name", "salary_usd"])
Hadoop = pd.DataFrame(columns = ["skill_name", "salary_usd"])
Java = pd.DataFrame(columns = ["skill_name", "salary_usd"])
Kubernetes = pd.DataFrame(columns = ["skill_name", "salary_usd"])
MLOps = pd.DataFrame(columns = ["skill_name", "salary_usd"])
MLP = pd.DataFrame(columns = ["skill_name", "salary_usd"])
PyTorch = pd.DataFrame(columns = ["skill_name", "salary_usd"])
Python = pd.DataFrame(columns = ["skill_name", "salary_usd"])
R = pd.DataFrame(columns = ["skill_name", "salary_usd"])
SQL = pd.DataFrame(columns = ["skill_name", "salary_usd"])
Scala = pd.DataFrame(columns = ["skill_name", "salary_usd"])
PySpark = pd.DataFrame(columns = ["skill_name", "salary_usd"])
Tableau = pd.DataFrame(columns = ["skill_name", "salary_usd"])
Tensorflow = pd.DataFrame(columns = ["skill_name", "salary_usd"])
```

```python
# Fill all the values in each newly created DataFrame
is_azure = salary_skills["Azure"].dropna()
azure["salary_usd"] = is_azure
azure["skill_name"] = "Azure"
is_docker = salary_skills["Docker"].dropna()
docker["salary_usd"] = is_docker
docker["skill_name"] = "Docker"
is_GCP = salary_skills["GCP"].dropna()
GCP["salary_usd"] = is_GCP
GCP["skill_name"] = "GCP"
is_Git = salary_skills["Git"].dropna()
Git["salary_usd"] = is_Git
Git["skill_name"] = "Git"
is_Hadoop = salary_skills["Hadoop"].dropna()
Hadoop["salary_usd"] = is_Hadoop
Hadoop["skill_name"] = "Hadoop"
is_Java = salary_skills["Java"].dropna()
Java["salary_usd"] = is_Java
Java["skill_name"] = "Java"
is_Kubernetes = salary_skills["Kubernetes"].dropna()
Kubernetes["salary_usd"] = is_Kubernetes
Kubernetes["skill_name"] = "Kubernetes"
is_MLOps = salary_skills["MLOps"].dropna()
MLOps["salary_usd"] = is_MLOps
MLOps["skill_name"] = "MLOps"
is_MLP = salary_skills["MLP"].dropna()
MLP["salary_usd"] = is_MLP
MLP["skill_name"] = "MLP"
is_PyTorch = salary_skills["PyTorch"].dropna()
PyTorch["salary_usd"] = is_PyTorch
PyTorch["skill_name"] = "PyTorch"
is_Python = salary_skills["Python"].dropna()
Python["salary_usd"] = is_Python
Python["skill_name"] = "Python"
is_R = salary_skills["R"].dropna()
R["salary_usd"] = is_R
R["skill_name"] = "R"
is_SQL = salary_skills["SQL"].dropna()
SQL["salary_usd"] = is_SQL
SQL["skill_name"] = "SQL"
is_Scala = salary_skills["Scala"].dropna()
Scala["salary_usd"] = is_Scala
Scala["skill_name"] = "Scala"
is_PySpark = salary_skills["PySpark"].dropna()
PySpark["salary_usd"] = is_PySpark
PySpark["skill_name"] = "PySpark"
is_Tableau = salary_skills["Tableau"].dropna()
Tableau["salary_usd"] = is_Tableau
Tableau["skill_name"] = "Tableau"
is_Tensorflow = salary_skills["Tensorflow"].dropna()
Tensorflow["salary_usd"] = is_Tensorflow
Tensorflow["skill_name"] = "Tensorflow"
```

```python
# Concatenate to all value continiously into 1 DataFrame
salary_categorize = pd.concat(
    [
        category_salary,
           azure, 
           docker, 
           GCP,
           Git, 
           Hadoop,
           Java,
           Kubernetes,
           MLOps, 
           MLP,
           PyTorch, 
           Python, 
           R, 
           SQL,
           Scala,
           PySpark,
           Tableau, 
           Tensorflow
          ]
         )
```

```python
# Optimization by type changing to int from float
salary_categorize["salary_usd"] = salary_categorize["salary_usd"].astype(int)
```
